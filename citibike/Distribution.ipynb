```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def exp_normal(n):
    exp = []
    for i in range(n):
        sample_i = np.random.normal(0,1,size=i+1)
        exp_i = float(sum(sample_i))/(i+1)
        exp.append(exp_i)
    return exp
```

```python
exp = exp_normal(10000)
```

```python
plt.plot(range(10000), exp)
```

```python
def exp_cauchy(n):
    exp = []
    for i in range(n):
        sample_i = np.random.standard_cauchy(size=i+1)
        exp_i = float(sum(sample_i))/(i+1)
        exp.append(exp_i)
    return exp
```

```python
exp2=exp_cauchy(10000)
```

```python
plt.plot(range(10000), exp2)
```
