```python
def area(name):
    if name in brooklyn:
        return 'b'
    else:
        return 'm'
```

```python
df['start_area'] = df['start station name'].apply(lambda name: area(name))
df['end_area'] = df['end station name'].apply(lambda name: area(name))
```

```python
df[df['start_area'] == 'b']['end_area'].value_counts()
```

```python
b = df[df['start_area'] == 'b']['end_area'].value_counts()['b']
m = df[df['start_area'] == 'b']['end_area'].value_counts()['m']
```

```python
p = float(m)/(m+b)
```

```python
print 'The probability is {percent:.2%}'.format(percent=p)
```

```python
#We can also try to use latitude and logitude data to seperate brooklyn and manhattan
#I found three useful points to build two lines to seperate these two areas
#The points I used were: P1(40.709232, -73.975591), P2(40.700123,-74.013872), P3(40.748778, -73.959798)
```

```python
Image(filename='map.png',width=400, height=400) 
```

```python
#construct two lines
x1 = np.array([[40.709232,1], [40.700123,1]])
y1 = np.array([-73.975591,-74.013872])
l1 = np.linalg.solve(x1, y1)
```

```python
x2 = np.array([[40.709232,1], [40.748778,1]])
y2 = np.array([-73.975591,-73.959798])
l2 = np.linalg.solve(x2, y2)
```

```python
def area2(df,la,lon):
    if df[la]<40.709232:
        if np.dot([df[la],1],l1)-df[lon]<0:
            return 'b'
        else:
            return 'm'
    else:
        if np.dot([df[la],1],l2)-df[lon]<0:
            return 'b'
        else:
            return 'm'
```

```python
df['start_area2']=df.apply(lambda df: area2(df,'start station latitude','start station longitude'),axis=1)
```

```python
df['end_area2']=df.apply(lambda df: area2(df,'end station latitude','end station longitude'),axis=1)
```

```python
df[df['start_area'] == 'b']['end_area'].value_counts()
```

```python
b2 = df[df['start_area2'] == 'b']['end_area'].value_counts()['b']
m2 = df[df['start_area2'] == 'b']['end_area'].value_counts()['m']
```

```python
p2 = float(m2)/(m2+b2)
```

```python
print 'The probability is {percent:.2%}'.format(percent=p2)
```
